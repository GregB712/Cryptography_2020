```python
# Algorithm to encode letters to 5-bits, according to PINAKAS1
       
def ConvertionPINAKAS1_5bit(plaintext):
    
    PINAKAS1=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z','.','!','?','(',')','-']
    PINAKAS1BINARY=["00000","00001","00010","00011","00100","00101","00110","00111","01000","01001","01010","01011","01100","01101","01110","01111","10000","10001","10010","10011","10100","10101","10110","10111","11000","11001","11010","11011","11100","11101","11110","11111"]

        
    new_plaintext = ""                                                  #Uncomment if we need to return string
    #new_plaintext_list =[]                                               #Comment if we need to return string
    for letter in plaintext:
        for i in range(0,32):
            if letter==PINAKAS1[i]:
                new_plaintext = new_plaintext+PINAKAS1BINARY[i]         #Uncomment if we need to return string
                #new_plaintext_list.append(PINAKAS1BINARY[i])             #Comment if we need to return string
    #return new_plaintext_list                                            #Comment if we need to return string
    return new_plaintext                                                #Uncomment if we need to return string
```

```python
# Algorithm to decode 5-bits to letters, according to PINAKAS1
       
def ConvertionPINAKAS1_5bit_decode(plaintext):
    
    PINAKAS1=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z','.','!','?','(',')','-']
    PINAKAS1BINARY=["00000","00001","00010","00011","00100","00101","00110","00111","01000","01001","01010","01011","01100","01101","01110","01111","10000","10001","10010","10011","10100","10101","10110","10111","11000","11001","11010","11011","11100","11101","11110","11111"]

    new_plaintext = ""      
    for i in range (0,len(plaintext),5):
        temp=plaintext[i]+plaintext[i+1]+plaintext[i+2]+plaintext[i+3]+plaintext[i+4]
        for i in range(0,32):
            if temp==PINAKAS1BINARY[i]:
                new_plaintext=new_plaintext + PINAKAS1[i]
            
    return new_plaintext
```

```python
# The LFSR sequence gets created by adding all the "taps" and then performing modulo 2, to find if next bit is 0 or 1.

def LFSR(initState,seqLength,FeedbackTaps):
    
    sequence=initState
    for i in range(len(FeedbackTaps)):
        FeedbackTaps[i]=len(initState) - FeedbackTaps[i]
        
    for i in range(len(initState),seqLength):
        summ=0
        for j in FeedbackTaps:
            summ=summ + int(sequence[j])
            
        nextBit=summ%2
        sequence = sequence + str(nextBit)
        
        for i in range(len(FeedbackTaps)):
            FeedbackTaps[i]=FeedbackTaps[i] + 1
        
    return sequence
```

```python
# Message from readme.txt

message = 'i!))aiszwykqnfcyc!?secnncvch'    #28 letters, 140 bits
print("_______________________________________________")
print("\nEncrypted message: " + message)
print("_______________________________________________")
# 5 bit conversion

message = ConvertionPINAKAS1_5bit(message)
print("\n5-bit Conversion:  " + message)

# We will represent the FeedBack Function x^10 + x^9 + x^7 + x^6 + 1 as a list of all the exponents.
FeedbackTaps=[10,9,7,6]


# We know from the exercise description that:
# For the 'ab' : ENC('ab') = '.s'
# So, we know that ENC(0000000001) = 1101010010

ab = '0000000001'
abEncrypted = '1101010010'

# We are going to find the 10-bit initial state, since we have both the plaintext 'ab', and its ciphertext '.s'.
# We are going to find it doing bitwise addition between them.

initState=''
for i in range(len(ab)):
    if ab[i] == abEncrypted[i]:
        initState = initState + '0'
    else:
        initState = initState + '1'
        
print("\nSeed:              " + initState)

# Now, since we have the 10-bit initial state, we are going to use the FeedBack Function to generate an 140-bit sequence.

sequence = LFSR(initState,len(message),FeedbackTaps)
print("\nLFSR sequence:     " + sequence)

# Since we now have the LFSR sequence, we can perform bitwise addition to the initial encypted message, to find the plaintext.

decrypted=''
for i in range(len(message)):
    if message[i] == sequence[i]:
        decrypted = decrypted + '0'
    else:
        decrypted = decrypted + '1'
print("_______________________________________________")
print("\nDecrypted message: " + ConvertionPINAKAS1_5bit_decode(decrypted))
print("_______________________________________________")
```
