```python
import random
from numpy import log as ln
n = 500            # Number of times i will test the formula
smallest = 15000    # Smallest value that can be tested
biggest = 100000   # Biggest value that can be tested

print("No | The Chosen Number | Î£(1/p) >= lnlnx-ln2 | Validation")
print("--------------------------------------------------------")
for i in range(n):
    summ=0 
    validation = True
    x = random.randint(smallest,biggest)      # Pick a large number each time
    lnexpr = ln(ln(x)) - ln(2)                # Compute the right expression 
    for j in range (1,x):
        summ = summ + 1/j                     # Compute the left expression
        valid = summ>=lnexpr
    if (valid==False):                        # If even one is false, formula is invalid
        validation = False
        print("Formula is invalid")
        break

    print(str(i) + ") | For number: " + str(x) + " | " + str("%.5f" % summ) + " >= " + str("%.5f" % lnexpr) + " | " + str(valid))
        
```
